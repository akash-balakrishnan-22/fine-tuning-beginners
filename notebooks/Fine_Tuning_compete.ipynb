```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
    )
```

```python
system_prompt = "You are an intelligent routing assistant. Classify the user query into the correct department. Output ONLY the label (e.g., SALES_TEAM)."
```

```python
def formatting_prompts_func(examples):
    convos = []
    texts = []

    system_prompt = "You are an intelligent routing assistant. Classify the user query into the correct department. Output ONLY the label (e.g., SALES_TEAM)."

    for input_text, output_text in zip(examples["input"], examples["output"]):
        conversation = [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": input_text},
            {"role": "assistant", "content": output_text},
        ]

        text = tokenizer.apply_chat_template(conversation, tokenize=False, add_generation_prompt=False)
        texts.append(text)

    return {"text": texts}
```

```python
dataset = dataset.map(formatting_prompts_func, batched=True)
```

```python
dataset[0]
```

```python
dataset[0]['text']
```

```python
from trl import SFTTrainer
```

```python
from transformers import TrainingArguments
```

```python
from unsloth import is_bfloat16_supported
```

```python
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False,
    args = TrainingArguments(
        per_device_train_batch_size = 2, #OOM
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
    ),
)
```

```python
trainer_stats = trainer.train()
```

```python
test_data = [
    ("Can you give me a quote for 100 licenses", "SALES_TEAM"),
    ("We want to upgrade to an enterprise subscription", "SALES_TEAM"),
    ("The webhook is failing with a timeout error", "TECH_TEAM"),
    ("My dashboard shows a blank screen after login", "TECH_TEAM"),
    ("How do I rotate my API keys", "TECH_TEAM"),
    ("I was charged twice for my subscription", "SUPPORT_TEAM"),
    ("I cannot reset my password", "SUPPORT_TEAM"),
    ("My account is locked due to multiple login attempts", "SUPPORT_TEAM"),
    ("Is there a discount for annual billing", "SALES_TEAM"),
    ("The mobile app crashes when I open settings", "TECH_TEAM"),
]
```

```python

y_true = []
y_pred = []

FastLanguageModel.for_inference(model)

print("Running evaluation...")

print(f"{'EXPECTED':<20} | {'PREDICTED (RAW)':<20} | {'MATCH?'}")
print("-" * 60)

for query, expected_label in test_data:
    messages = [
        {"role": "system", "content": "You are an intelligent routing assistant. Classify the user query into the correct department. Output ONLY the label (e.g., SALES_TEAM)."},
        {"role": "user", "content": query},
    ]
    inputs = tokenizer.apply_chat_template(messages, tokenize=True, add_generation_prompt=True, return_tensors="pt").to("cuda")

    outputs = model.generate(input_ids=inputs, max_new_tokens=64, use_cache=True, temperature=0.01)

    prediction = tokenizer.decode(outputs[0][inputs.shape[1]:], skip_special_tokens=True)

    print(f"{expected_label:<20} | {repr(prediction):<20} | {expected_label == prediction.strip()}")
```
